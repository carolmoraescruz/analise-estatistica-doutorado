```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('ticks')
# sns.set(font_scale = 1.5)
```

## Gráficos unificados - 7, 28 e 91 dias

```python
mea_etapa3_port = pd.read_csv('mea_etapa3_port.csv')
rfta_etapa3_port = pd.read_csv('rfta_etapa3_port.csv')
rcta_etapa3_port = pd.read_csv('rcta_etapa3_port.csv')
rcta_etapa3_port
```

```python
sns.catplot(
    data=mea_etapa3_port,
    kind="bar",
    x="Fórmula",
    y="MEA",
    hue="Dias",
    ci="sd",
    palette="Greys",
    alpha=0.7,
    capsize=0.1,
    errwidth=0.8,
    height=5,
    aspect=1.5,
)
plt.ylabel("DA (g/cm³)")
plt.ylim([1.8, 2.3])
plt.savefig("MEA completo etapa 3.svg")
plt.show()
```

```python
sns.catplot(
    data=rfta_etapa3_port,
    kind="bar",
    x="Fórmula",
    y="RFTA",
    hue="Dias",
    ci="sd",
    palette="Greys",
    alpha=0.7,
    capsize=0.1,
    errwidth=0.8,
    height=5,
    aspect=1.5,
)
plt.ylabel("RFTA (MPa)")
plt.savefig("RFTA completo etapa 3.svg")
plt.show()
```

```python
sns.catplot(
    data=rcta_etapa3_port,
    kind="bar",
    x="Fórmula",
    y="RCTA",
    hue="Dias",
    ci="sd",
    palette="Greys",
    alpha=0.7,
    capsize=0.1,
    errwidth=0.8,
    height=5,
    aspect=1.5,
)
plt.ylabel("RCTA (MPa)")
plt.savefig("RCTA completo etapa 3.svg")
plt.show()
```

## Gráficos do artigo (inglês) - 7, 28 e 91 dias

```python
mea_etapa3_ingles = pd.read_csv("MEA_etapa3_ingles.csv")
rfta_etapa3_ingles = pd.read_csv("RFTA_etapa3_ingles.csv")
rcta_etapa3_ingles = pd.read_csv("RCTA_etapa3_ingles.csv")
mea_etapa3_ingles
```
